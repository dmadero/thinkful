```python
import numpy as np
import pandas as pd
import scipy
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def graphs(distr) :
    plt.hist(distr, bins=20, color='c')
    # Add a vertical line at the mean.
    plt.axvline(distr.mean(), color='b', linestyle='solid', linewidth=2)
    # Add a vertical line at one standard deviation above the mean.
    plt.axvline(distr.mean() + distr.std(), color='b', linestyle='dashed', linewidth=2)
    # Add a vertical line at one standard deviation below the mean.
    plt.axvline(distr.mean()-distr.std(), color='b', linestyle='dashed', linewidth=2) 
    state = 'The Standard Deviation is {} and the Mean is {}'
    return plt.show() or print(state.format(distr.std(),distr.mean()))
```

```python
laplace = np.random.laplace([3.5,-1,3019], 100)
graphs(laplace)
```

```python
rayleigh = np.random.rayleigh(200,100)
graphs(rayleigh)
```

```python
logi = np.random.logistic(18,3,100)
graphs(logi)
```

```python
uni = np.random.uniform(-12,43,100)
graphs(uni)
```

```python
vmis = np.random.vonmises(2,4,100)
graphs(vmis)
```

```python
stdrex = np.random.standard_exponential(100)
graphs(stdrex)
```

```python
norm1 = np.random.normal(5,.5,100)
norm2 = np.random.normal(10,1,100)
mistery = norm1+norm2
graphs(mistery)
```

```python
pop1 = np.random.binomial(10,.5,100)
pop2 = np.random.binomial(10,.2,100)
graphs(pop1)
graphs(pop2)
```

When the populations are increased, the standard deviation will increase similar while de mean will remain fairly similar.

```python
pop3 = np.random.binomial(10,.5,1000)
pop4 = np.random.binomial(10,.2,1000)
graphs(pop3)
graphs(pop4)
```

```python
pop5 = np.random.binomial(10,.5,20)
pop6 = np.random.binomial(10,.2,20)
graphs(pop5)
graphs(pop6)
```

```python
#The mean will decrease while the stddev will remain fairly similar
from scipy.stats import ttest_ind
pop7 = np.random.binomial(10,.3,100)
sample1 = np.random.choice(pop7, 15, replace=True)
sample2 = np.random.choice(pop7, 25, replace=True)
print(ttest_ind(sample2, sample1, equal_var=False))
```
