```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
the_data = pd.read_csv('/Users/Dan/AnacondaProjects/Files/causes_of_death.csv')
df = pd.DataFrame(the_data)
df
```

```python
newdata1 = the_data.loc[the_data.State!='United States', ['Year', 'Cause Name', 'State', 'Deaths']]
newdata1.head(1)
```

```python
Deaths_TX = newdata1[(newdata1.State == 'Texas')]
Deaths_TX.rename(columns = {'Cause Name':'Cause'})
```

```python
plt.scatter(
    x = Deaths_TX['Year'], 
    y = Deaths_TX['Deaths'],
    )
plt.ylabel('Deaths')
plt.xlabel('Years')
plt.show()
```

```python
plt.scatter(y = Deaths_TX.Deaths <= 25000, x = Deaths_TX['Year'])
plt.show()
```

```python
plt.plot(Deaths_TX.Deaths)
plt.show()
```

```python
plt.hist(Deaths_TX.Deaths)
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(Deaths_TX['Deaths'], color='purple')
plt.ylabel('Year')
plt.title('Subplot')

plt.subplot(1, 2, 2)
plt.plot(Deaths_TX['Year'], color='green')
plt.ylabel('Year')
plt.title('Subplot_2')
plt.show()
```
